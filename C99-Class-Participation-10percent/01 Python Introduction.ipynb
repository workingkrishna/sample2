# Introduction to Python
## What is Python?
- Object oriented language
- Interpreted language
- Supports dynamic data type
- Independent of platforms
- Simple and easy grammar
- High-level internal object data types
- Automatic memory management
- It’s free (open source)!

## Why use Python?
- Fun-to-use "Scripting language"
- Object-oriented
- 8 out of 10 US universities use it
- Very easy to learn
- Powerful, scalable, easy to maintain
- Lots of libraries

## History of Python

## Community Support for Python

- [PyData](https://groups.google.com/g/pydata): A Google Group list for questions related to Python for data analysis and pandas

- [pystatsmodels](https://www.statsmodels.org/stable/index.html): For statsmodels or pandas-related questions

- [numpy-discussion](https://numpy.org/community/): For NumPy-related questions


## Conferences
Many conferences have financial support available for those who cannot afford admission or travel to the conference.

- [USPyCon](https://us.pycon.org/2022/) and [EuroPython](https://www.europython-society.org/): The two main general Python conferences in North America and Europe, respectively

- [pydata](https://pydata.org/): A worldwide series of regional conferences targeted at data science and data analysis use cases

- International and regional [PyCon](https://pycon.org/) conferences
